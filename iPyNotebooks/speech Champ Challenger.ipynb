```python
from scipy import stats  
from statsmodels.formula.api import ols
dep2 = ['ACC', 'CC',
       'RR',
       'ACRR']
data2 = data2.dropna()
for y in dep2:
    print "OLS for" + y
    formula = y + '~ SvA + SOFT + SvA * SOFT'
    model = ols(formula,data=data2).fit()
    print model.summary()
```
