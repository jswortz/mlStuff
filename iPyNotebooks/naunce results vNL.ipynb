```python
import pandas as pd
import xlrd
from difflib import SequenceMatcher
from functools import partial


#defining the similarity between two word - used for the 
def similar(a, b):
    try:
        return SequenceMatcher(None, a, b).ratio()
    except:
        return 0

data = pd.read_excel('C:\\Users\\jswortz\\Documents\\nlspeech.xlsx',header=0, sheetname='nlspeech.csv', )
data['Transcription'] = data.Transcription.fillna(method = 'pad')  #this propegates the transcription values from the x values

#data.Transcription ---It worked for the forward prop
#data.describe()
example = 'zero zero [bg noise]'

def cleanText(x):
    z = x.split('[')
    try:
        w = z[1].split(']')
        return z[0] + w[1]
    except:
        return x
    
data['Transcription'] = data.apply(lambda row: cleanText(row['Transcription']), axis=1)


data.columns.values
data['Similarity']=data.apply(lambda row: similar(row['SPOK'], row['Transcription']), axis=1)

#NOT SURE IF WE SHOULD USE SPOK OR RAWT
print data.columns.values
print data.Similarity.describe()
print
data.LA.fillna(0)
#print data.LA
print
#print data.Group.unique()
```
