## Guide

### Note: API Documentation here: ###

```python
import json
import requests
import urllib2
import pandas as pd
url = "https://apps.westvacd.com/vacdcentral/api/clients/212/skills"

req = requests.get(url)
skills = pd.DataFrame(req.json()['Skill'])
skills
```

### Definition of the function that pulls interval data by skill

```python
def getSkillData(skill):
    url = "https://apps.westvacd.com/vacdcentral/api/skill_interval"
    data = json.dumps({"skill_numbers": [skill], 
                       "start_time": "2016-06-01T00:00:00-0500", 
                       "end_time": "2016-09-20T23:59:59-0500",
               
                       
                       "short_abandoned_threshold": "0"})
    headers = {"Accept" : "application/json","Content-type": "application/json"}
    req = urllib2.Request(url, data, headers)
    response = urllib2.urlopen(req)
    return response
# response = getSkillData(3101)
# j = json.load(response)
# data = pd.DataFrame.from_dict(j['Intervals']['IntervalData'])
# data
# data.to_csv("VandyData.csv", encoding='utf-8')
```

### Loop through the skills, write data to a csv file
## Note: Change the CSV location, as it will append to an existing location
## Note2: Will also need to copy column headers manually

```python
df = pd.DataFrame()

for i in range(0, len(skills)-1):
    r = skills.ix[[i]]
    response = getSkillData(int(r.iloc[0]['SkillNumber']))
    j = json.load(response)
    d = pd.DataFrame.from_dict(j['Intervals']['IntervalData'])
    d2 = r.append(d).fillna(method='ffill')
    d3 = d2[d2.CallsOffered > 0]
    with open('VandyData.csv', 'a') as f:
        d3.to_csv(f, header=False, encoding='utf-8')
```
